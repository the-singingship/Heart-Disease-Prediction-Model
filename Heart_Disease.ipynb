```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df= pd.read_csv('heart_disease_data.csv')
```

1: Defective Heart   0:Healthy Heart

```python
x = df.drop(columns='target', axis=1)
```

```python
x
```

```python
y=df['target']
```

```python
y
```

Splitting dataset into train and test

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,stratify=y, random_state=23)
```

```python
print(x_test.shape, x_train.shape)
```

Model Training: Logistic regression (Classification problem)

```python
model = LogisticRegression()
```

```python
#Train the modle with training data
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_train
```

```python
score= accuracy_score(y_test,y_pred)
```

```python
print('Accuracy score:', score)
```

Building a Predictive System

```python
input_data=(62,0,0,140,268,0,0,160,0,3.6,0,2,2)

#Change the input_data into array

arr= np.array(input_data)
```

```python
print(arr)
```

```python
prediction = model.predict(arr.reshape(1,-1))
```

```python
print(prediction)
```

```python
if prediction==0:
    print('The person does not have heart disease. They are healthy')
else:
    print('The person has heart disease. Please consult a doctor')
```
